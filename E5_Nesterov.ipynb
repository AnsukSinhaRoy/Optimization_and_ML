```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
alpha = 0.1
beta = 0.9
max_iters = 50
x_mom = 5
x_nag = 5
v_mom = 0
v_nag = 0

# Initialize history arrays
history_mom = np.zeros(max_iters)
history_nag = np.zeros(max_iters)

# Optimization loop
for k in range(max_iters):
    # Momentum update
    grad_mom = 2 * x_mom
    v_mom = beta * v_mom - alpha * grad_mom
    x_mom = x_mom + v_mom
    history_mom[k] = x_mom
    
    # Nesterov's Accelerated Gradient (NAG) update
    y_nag = x_nag + beta * v_nag
    grad_nag = 2 * y_nag
    v_nag = beta * v_nag - alpha * grad_nag
    x_nag = x_nag + v_nag
    history_nag[k] = x_nag

# Plotting
plt.figure()
plt.plot(range(1, max_iters + 1), history_mom, 'g-o', linewidth=1.5, label='Momentum')
plt.plot(range(1, max_iters + 1), history_nag, 'b-s', linewidth=1.5, label='Nesterov')
plt.xlabel('Iteration')
plt.ylabel('x value')
plt.title('Momentum vs. Nesterov’s Accelerated Gradient Descent')
plt.legend()
plt.grid(True)
plt.show()
```
